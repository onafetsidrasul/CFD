```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('yy.csv')
```

```python
print(df.head())
```

```python
def u(y):
    return -(y / 0.05) * (-4.9932) * (0.5 - y / 2)
```

```python
y_vals = np.linspace(0, 1, 100)
u_vals = u(y_vals)
```

```python
plt.plot(df['Points:1'], df['U:0'], marker='o', linestyle='None', label='simulation')
plt.plot(y_vals, u_vals, label='theory')
plt.xlabel('y')
plt.ylabel('U')
plt.title('U development')
plt.legend()
```
